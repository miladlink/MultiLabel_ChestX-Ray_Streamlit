```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch
from torch import nn
from torch.utils.data import DataLoader, random_split
from torchvision import transforms, models

from dataset import NIHDataset
from utils import *
from train import *
```

```python
# load the labels from csv file
labels = pd.read_csv ('../NIH/sample_labels.csv')
labels.head (5)
```

```python
pathology_list = ['Cardiomegaly', 'Emphysema', 'Effusion', 'Hernia', 'Nodule', 'Pneumothorax', 'Atelectasis',
                  'Pleural_Thickening', 'Mass','Edema', 'Consolidation', 'Infiltration', 'Fibrosis', 'Pneumonia']

# encode pathologies
for pathology in pathology_list :
    labels [pathology] = labels ['Finding Labels'].apply (lambda x: 1 if pathology in x else 0)

labels ['No Findings'] = labels ['Finding Labels'].apply (lambda x: 1 if 'No Finding' in x else 0)
labels.head (5)
```

```python
# drop additional features
labels = labels.drop (list (labels.iloc [:,1:11].columns.values), axis = 1)
labels.head (5)
```

create dataset

```python
# transforms for dataset
tfms = transforms.Compose ([
    transforms.RandomRotation ((-20,+20)),
    transforms.Resize ((224,224)),
    transforms.ToTensor (),
    transforms.Normalize (mean = [0.485, 0.456, 0.406], 
                          std = [0.229, 0.224, 0.225])
])
```

```python
imgs_dir = '../NIH/sample/images'
```

```python
dataset = NIHDataset (imgs_dir, labels, transform = tfms)
```

```python
# show random picture of dataset
index = np.random.choice (len (dataset))
img, lbl = dataset [index]
print (img.shape)
class_names = list (np.where (lbl == 1)[0])
for class_name in class_names:
    print (pathology_list [class_name], end = ' ')
imshow (img)
```

```python
# shuffle & split dataset to train, valid, test
train_ds, valid_ds, test_ds = random_split(dataset, [5000,303,303])
print ('Train =', len (train_ds))
print ('Valid =', len (valid_ds))
print ('Test =', len (test_ds))
```

```python
# create dataloader
train_dl = DataLoader(train_ds, batch_size = 64, shuffle = True)
valid_dl = DataLoader(valid_ds,batch_size = 64, shuffle = False)
test_dl = DataLoader(test_ds, batch_size = 64, shuffle = False)
```

```python
device = torch.device ('cuda' if torch.cuda.is_available () else 'cpu')
```

```python
# load pretrained model
model = models.mobilenet_v2 (pretrained = True)
```

```python
#freaze weights
for param in model.parameters():
    param.requires_grad = False

# change top layers
model.classifier = nn.Sequential(
    nn.Linear (1280, 512),
    nn.ReLU (),
    nn.Dropout (0.5),
    nn.Linear(512, 14),
    nn.Sigmoid()
)
```

```python
# testing the model
img, lbl = next (iter (train_dl))
img = img.to(device)
lbl = lbl.to(device)
output = model.to(device) (img)
print (output.shape)
```
